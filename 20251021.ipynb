```python
mapbiomas_dir = '/projectnb/modislc/users/chishan/data/MapBiomas/MAPBIOMAS'
#read the csv file
import pandas as pd
mapbiomas_df = pd.read_csv(f'{mapbiomas_dir}/samples-BRAZIL.csv')
mapbiomas_df.head()
mapbiomas_df[mapbiomas_df['year']>2000]

# Cell: filter rows where reference == 15 (robust to column name)
import pandas as pd
# ensure mapbiomas_df exists in the notebook namespace
if 'mapbiomas_df' not in globals():
    raise NameError('mapbiomas_df not found. Run the cell that loads the CSV first.')
# possible column names to consider for 'reference'
ref_candidates = ['reference', 'ref', 'Reference', 'REF']
ref_col = next((c for c in ref_candidates if c in mapbiomas_df.columns), None)
if ref_col is None:
    raise KeyError(f'No reference column found. Candidates tried: {ref_candidates}. Columns: {list(mapbiomas_df.columns)}')
df_ref15 = mapbiomas_df[mapbiomas_df[ref_col] == 15].copy()
print(f'Found {len(df_ref15)} rows with {ref_col} == 15')
# show first rows and the columns (for inspection)
df_ref15 = df_ref15[df_ref15['year'] > 2008]
# df_sample = df_ref15.sample(n=300, random_state=42)

file_name = "/projectnb/modislc/users/chishan/data/MapBiomas/Limites_Amazonia_Legal_2024_shp/Limites_Amazonia_Legal_2024.shp"
amazonas = gpd.read_file(file_name)

gdf = df_ref15.to_crs(amazonas.crs)
amazonas_union = amazonas.unary_union
gdf_amazon = gdf[gdf.geometry.intersects(amazonas_union)].reset_index(drop=True)

sample_size_amazon = min(len(gdf_amazon), 100)  # limit draw to available features in Amazon area
gdf_amazon_sample = gdf_amazon.sample(n=sample_size_amazon, random_state=42).reset_index(drop=True)
gdf_amazon_sample["sample_id"] = gdf_amazon_sample.index + 1
# gdf_amazon_sample

# Save the sampled GeoDataFrame as an ESRI Shapefile (attributes like 'year' are preserved)
out_shp = "/Users/samzhang/Downloads/mapbiomas_amazon_sample_100.shp"
# Driver is inferred from the .shp extension; no need to pass driver explicitly
# Note: Shapefile field names are limited to 10 characters and may be truncated.
gdf_amazon_sample.to_file(out_shp)
print(f"Wrote shapefile to: {out_shp}")
```
