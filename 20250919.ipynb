```python
# Enhanced heatmap highlighting small area clusters (占地面积很小)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import math
import pandas as pd
from pathlib import Path

# Check if CSV files exist locally (if exported from Drive to local folder)
csv_dir = Path('/projectnb/modislc/users/chishan/data/Glance')  # Adjust path as needed
csv_files = {
    50: csv_dir / 'cluster_pixel_stats_50cl.csv',
    70: csv_dir / 'cluster_pixel_stats_70cl.csv', 
    100: csv_dir / 'cluster_pixel_stats_100cl.csv'
}

# Alternative: if files are in current directory
current_dir = Path('.')
for cluster_count in [50, 70, 100]:
    local_file = current_dir / f'cluster_pixel_stats_{cluster_count}cl.csv'
    if local_file.exists():
        csv_files[cluster_count] = local_file
        
df_50 = pd.read_csv(csv_files[50])
# Sort by percentage descending
sorted_df = df_50.sort_values('percentage', ascending=False)

# Configuration and important clusters (80% coverage clusters)
ANALYSIS_CONFIG = {'n_clusters': 50}
K = ANALYSIS_CONFIG['n_clusters']
important_clusters = [48, 24, 1, 34, 22, 38, 10, 49, 36, 35, 47, 2, 18, 45, 12, 15, 20, 44, 6, 33, 27, 14, 5, 13, 43, 39, 32, 8, 46, 17, 41]

# Load the full training data
df_full = pd.read_csv(f'/projectnb/modislc/users/chishan/data/Glance/training_with_clusters_{K}cl.csv')

# Filter to only important clusters
df_clusters_filtered = df_full[df_full['cluster'].isin(important_clusters)].copy()

# Create data for ALL clusters (both important and small area clusters)
df_all_clusters = df_full.groupby(['LC_Class', 'cluster']).size().reset_index(name='count')

# Create pivot table for ALL clusters
pivot_all = df_all_clusters.pivot(index='LC_Class', columns='cluster', values='count').fillna(0)

# Reorder columns based on percentage descending (from sorted_df)
cluster_order = sorted_df['cluster_id'].tolist()
# Only keep clusters that exist in our pivot table
cluster_order = [c for c in cluster_order if c in pivot_all.columns]
# Add any remaining clusters that might not be in the percentage data
remaining_clusters = [c for c in pivot_all.columns if c not in cluster_order]
cluster_order.extend(remaining_clusters)
# Reorder the pivot table columns
pivot_all = pivot_all[cluster_order]

# Create two separate heatmaps: one comprehensive, one highlighting small area clusters
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 12))

# 1. COMPREHENSIVE HEATMAP - All clusters with important ones highlighted
# Create a mask for small area clusters
all_clusters = set(range(K))  # Assuming clusters are 0 to K-1
small_area_clusters = all_clusters - set(important_clusters)

# Create annotation array with special marking for small area clusters
annot_array = pivot_all.values.copy()
annot_labels = []
for i in range(pivot_all.shape[0]):
    row_labels = []
    for j in range(pivot_all.shape[1]):
        cluster_id = pivot_all.columns[j]
        value = pivot_all.values[i, j]
        if cluster_id in small_area_clusters and value > 0:
            # Mark small area clusters with asterisk
            row_labels.append(f"{int(value)}*")
        else:
            row_labels.append(f"{int(value)}" if value > 0 else "0")
    annot_labels.append(row_labels)

# Plot comprehensive heatmap
sns.heatmap(pivot_all, annot=annot_labels, fmt='', cmap='YlOrRd', 
            ax=ax1, cbar_kws={'label': 'Number of samples'})
ax1.set_title(f'All Clusters: LC Class vs Cluster Distribution (Sorted by % Coverage Descending)\n(* = Small area clusters, {len(small_area_clusters)} clusters outside 80% coverage)', 
              fontsize=14)
ax1.set_xlabel('Cluster ID (Ordered by Coverage %)')
ax1.set_ylabel('LC Class')

# Add vertical lines to separate important from small area clusters
for cluster_id in small_area_clusters:
    if cluster_id in pivot_all.columns:
        cluster_pos = list(pivot_all.columns).index(cluster_id)
        ax1.axvline(x=cluster_pos, color='red', linestyle='--', alpha=0.3, linewidth=1)

# 2. FOCUSED HEATMAP - Only small area clusters (maintain percentage-based order)
small_area_in_order = [c for c in cluster_order if c in small_area_clusters]
pivot_small_area = pivot_all[small_area_in_order]

if pivot_small_area.shape[1] > 0:  # Only plot if there are small area clusters with data
    sns.heatmap(pivot_small_area, annot=True, fmt='g', cmap='Reds', 
                ax=ax2, cbar_kws={'label': 'Number of samples'})
    ax2.set_title(f'Small Area Clusters Only (Sorted by % Coverage) - {len(small_area_clusters)} clusters representing ~20% of area', 
                  fontsize=14)
    ax2.set_xlabel('Small Area Cluster ID (Ordered by Coverage %)')
    ax2.set_ylabel('LC Class')
else:
    ax2.text(0.5, 0.5, 'All clusters with data are important clusters', 
             ha='center', va='center', transform=ax2.transAxes, fontsize=14)
    ax2.set_title('No Data in Small Area Clusters')

plt.tight_layout()
plt.show()

# Statistical summary of data distribution
print("=== DATA DISTRIBUTION ANALYSIS ===")
total_samples = df_all_clusters['count'].sum()
important_samples = df_clusters_filtered.shape[0]
non_important_samples = total_samples - important_samples

print(f"Total training samples: {total_samples}")
print(f"Samples in important clusters (80% area coverage): {important_samples} ({important_samples/total_samples*100:.1f}%)")
print(f"Samples in small area clusters (20% area coverage): {non_important_samples} ({non_important_samples/total_samples*100:.1f}%)")

# Analyze small area cluster data if it exists
if non_important_samples > 0:
    df_small_area = df_full[df_full['cluster'].isin(small_area_clusters)]
    small_area_grouped = df_small_area.groupby(['LC_Class', 'cluster']).size().reset_index(name='count')
    
    print(f"\n=== SMALL AREA CLUSTERS ANALYSIS ===")
    print(f"Number of small area clusters with data: {df_small_area['cluster'].nunique()}")
    print(f"LC classes in small area clusters: {sorted(df_small_area['LC_Class'].unique())}")
    
    # Top clusters by sample count in small area clusters
    cluster_counts = df_small_area['cluster'].value_counts().head(5)
    print(f"\\nTop 5 small area clusters by sample count:")
    for cluster_id, count in cluster_counts.items():
        print(f"  Cluster {cluster_id}: {count} samples")
    
    # LC classes most affected by exclusion of small area clusters
    lc_comparison = []
    for lc_class in df_full['LC_Class'].unique():
        total_lc = len(df_full[df_full['LC_Class'] == lc_class])
        important_lc = len(df_clusters_filtered[df_clusters_filtered['LC_Class'] == lc_class])
        loss_pct = (total_lc - important_lc) / total_lc * 100
        lc_comparison.append({
            'LC_Class': lc_class,
            'total_samples': total_lc,
            'important_samples': important_lc,
            'non_important_samples': total_lc - important_lc,
            'loss_percentage': loss_pct
        })
    
    lc_comp_df = pd.DataFrame(lc_comparison).sort_values('loss_percentage', ascending=False)
    
    print(f"\\n=== LC CLASSES MOST AFFECTED BY FOCUSING ON IMPORTANT CLUSTERS ===")
    print("LC classes with highest sample loss when excluding small area clusters:")
    print(lc_comp_df.head(10)[['LC_Class', 'total_samples', 'important_samples', 'loss_percentage']])
    
    # Recommendations
    print(f"\\n=== RECOMMENDATIONS ===")
    high_loss_classes = lc_comp_df[lc_comp_df['loss_percentage'] > 30]['LC_Class'].tolist()
    if high_loss_classes:
        print(f"1. HIGH PRIORITY - These LC classes lose >30% of samples when focusing on important clusters:")
        for lc_class in high_loss_classes[:5]:
            print(f"   • LC {lc_class}: {lc_comp_df[lc_comp_df['LC_Class']==lc_class]['loss_percentage'].iloc[0]:.1f}% sample loss")
        print(f"   → Consider collecting additional samples for these classes in important clusters")
    
    print(f"\\n2. Consider whether small area clusters contain unique environmental conditions")
    print(f"   that might be important for your specific research objectives")
    
    print(f"\\n3. The 80% area coverage approach focuses on spatial representativeness,")
    print(f"   but some rare LC classes might be better represented in smaller clusters")

else:
    print("\\nAll training data is located within important clusters!")
    print("No additional data collection needed in small area clusters.")
```

```python
import geemap
import ee
import matplotlib.pyplot as plt
import numpy as np

# Initialize Earth Engine
ee.Initialize(project='ee-zcs')

# Create interactive map
Map = geemap.Map()
Map.add_basemap('HYBRID')

# ============================================================================
# VISUALIZATION PARAMETERS
# ============================================================================

# Land cover classes and colors
lc_classes = ['Water', 'Snow/Ice', 'Developed', 'Bare', 'Forest', 'Shrub', 'Herbaceous']
lc_palette = ['#386cb0', 'white', 'red', 'black', '#336600', '#28A765', '#ffff99']
lc_viz = {'min': 1, 'max': 7, 'palette': lc_palette}

# Uncertainty visualization
uncertainty_viz = {
    'palette': ['#440154', '#3b528b', '#21918c', '#5ec962', '#fde725'],
    'min': 0.2,
    'max': 0.75
}

# ============================================================================
# LOAD DATA
# ============================================================================

# Define region (Africa example - adjust as needed)
africa_boundary = ee.FeatureCollection("projects/GLANCE/ANCILLARY/CONTINENTS_EXPORT/GLANCE_V01_AF_PROJ_LAND_buf")
region = africa_boundary.geometry()

# GLANCE dataset (2020)
glance_2020 = ee.ImageCollection('projects/GLANCE/DATASETS/GLANCE_V01_AF') \
    .filterDate('2020-01-01', '2021-01-01') \
    .mosaic() \
    .select('.*2020.*')

# Google embeddings
embeddings = ee.ImageCollection('projects/google/embeddings/v1') \
    .filterDate('2020-01-01', '2021-01-01') \
    .filterBounds(region) \
    .mosaic()

# Training data (assuming these exist in your workspace)
training_data = ee.FeatureCollection('projects/your-project/training_data')

# ============================================================================
# RANDOM FOREST CLASSIFIER WITH UNCERTAINTY
# ============================================================================

def random_forest_uncertainty(embeddings_img, training_fc, class_property='Glance_Class_ID_level1'):
    """
    Train Random Forest classifier and compute uncertainty

    Args:
        embeddings_img: Input embeddings image
        training_fc: Training feature collection
        class_property: Name of class property

    Returns:
        Tuple of (classified_image, uncertainty_image)
    """

    # Get band names
    band_names = embeddings_img.bandNames()

    # Train classifier with probability output
    classifier = ee.Classifier.smileRandomForest(
        numberOfTrees=100,
        minLeafPopulation=5,
        bagFraction=0.7
    ).train(
        features=training_fc.filter(ee.Filter.notNull(band_names)),
        classProperty=class_property,
        inputProperties=band_names
    ).setOutputMode('MULTIPROBABILITY')

    # Classify image
    probability_image = embeddings_img.classify(classifier)

    # Compute uncertainty (1 - max probability)
    max_probability = probability_image.arrayReduce(ee.Reducer.max(), [0]).arrayGet([0])
    uncertainty = ee.Image(1.0).subtract(max_probability).rename('uncertainty')

    return probability_image, uncertainty

# ============================================================================
# MINIMUM DISTANCE CLASSIFIERS
# ============================================================================

def minimum_distance_classification(embeddings_img, training_fc,
                                  distance_type='cosine',
                                  class_property='Glance_Class_ID_level1'):
    """
    Perform minimum distance classification

    Args:
        embeddings_img: Input embeddings image
        training_fc: Training feature collection
        distance_type: 'cosine' or 'euclidean'
        class_property: Name of class property

    Returns:
        Classified image
    """

    # Get band names
    band_names = embeddings_img.bandNames()

    # Create classifier
    classifier = ee.Classifier.minimumDistance(distance_type).setOutputMode('REGRESSION')

    # Train classifier
    trained_classifier = classifier.train(
        features=training_fc.filter(ee.Filter.notNull(band_names)),
        classProperty=class_property,
        inputProperties=band_names
    )

    # Classify image
    classified = embeddings_img.classify(trained_classifier)

    return classified

# ============================================================================
# VISUALIZATION FUNCTIONS
# ============================================================================

def style_training_points(fc, class_property='Glance_Class_ID_level1'):
    """
    Style training points based on land cover class

    Args:
        fc: Feature collection
        class_property: Name of class property

    Returns:
        Styled feature collection
    """

    def style_feature(feature):
        lc_class = ee.Number(feature.get(class_property))
        color = ee.Algorithms.If(
            lc_class.eq(1), '#386cb0',  # Water
            ee.Algorithms.If(
                lc_class.eq(3), 'red',  # Developed
                ee.Algorithms.If(
                    lc_class.eq(4), 'black',  # Bare
                    ee.Algorithms.If(
                        lc_class.eq(5), '#336600',  # Forest
                        ee.Algorithms.If(
                            lc_class.eq(6), '#28A765',  # Shrub
                            '#ffff99'  # Herbaceous (default)
                        )
                    )
                )
            )
        )

        return feature.set('style', {
            'color': color,
            'pointSize': 6
        })

    return fc.map(style_feature)

# ============================================================================
# MAIN ANALYSIS
# ============================================================================

print("=== GLANCE Land Cover Analysis with Uncertainty ===")

# Add base layers
Map.centerObject(region, 6)
Map.addLayer(glance_2020, lc_viz, 'GLANCE 2020 Land Cover', True)

# Perform Random Forest classification with uncertainty
print("\\nTraining Random Forest classifier...")
rf_classified, uncertainty = random_forest_uncertainty(embeddings, training_data)

# Add uncertainty layer
Map.addLayer(uncertainty, uncertainty_viz, 'Uncertainty Hotspots', False)

# Perform minimum distance classifications
print("\\nPerforming minimum distance classifications...")
min_dist_cosine = minimum_distance_classification(embeddings, training_data, 'cosine')
min_dist_euclidean = minimum_distance_classification(embeddings, training_data, 'euclidean')

# Add classification results
Map.addLayer(min_dist_cosine, {'min': 0.2, 'max': 1.0}, 'Min Distance (Cosine)', False)
Map.addLayer(min_dist_euclidean, {'min': 0.2, 'max': 1.0}, 'Min Distance (Euclidean)', False)

# Style and add training points
styled_training = style_training_points(training_data)
Map.addLayer(styled_training, {}, 'Training Points', False)

# ============================================================================
# UNCERTAINTY ANALYSIS
# ============================================================================

# Extract uncertainty values at training locations
training_uncertainty = uncertainty.reduceRegions(
    collection=training_data,
    reducer=ee.Reducer.first(),
    scale=10,
    tileScale=2
).filter(ee.Filter.notNull(['first']))

# Identify high uncertainty samples (> 0.7)
high_uncertainty = training_uncertainty.filter(ee.Filter.gt('first', 0.7))
styled_high_uncertainty = style_training_points(high_uncertainty)

Map.addLayer(styled_high_uncertainty, {}, 'High Uncertainty Samples', False)

print(f"\\nAnalysis complete!")
print(f"High uncertainty samples: {high_uncertainty.size().getInfo()}")

# Display the map
Map
```
